```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data using a dummy function
X = np.arange(15)
Y = (X-5)**2 + 3
print(X,Y)
```

```python
# Plot the curve
plt.style.use("seaborn")
plt.plot(X,Y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Implementing Gradient Descent
# Assuming y = (x-5)^2 + 3 so gradient descent = differentiation (y) = 2(x-5)+0
# initialize x with any random value
x=15
y=0
learningRate=0.1

for i in range(50):
    gradient = 2*(x-5)
    x=x-learningRate*gradient
    y=(x-5)**2+3
    plt.scatter(x,y)
    print(x)
plt.plot(X,Y)
plt.show()
```
